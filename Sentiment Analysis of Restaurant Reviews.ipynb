```python
from textblob import TextBlob  # Using TextBlob for sentiment analysis

# Load the restaurant reviews dataset (assuming it's in a CSV file)
reviews = pd.read_csv("cleanned_data.csv")
reviews["reviews_list"] = reviews["reviews_list"].astype(str)
# Preprocess the text:
reviews["cleaned_text"] = reviews["reviews_list"].apply(lambda text: text.lower().replace("[^a-zA-Z]", " ").strip())

# Analyze sentiment using TextBlob:
reviews["sentiment"] = reviews["cleaned_text"].apply(lambda text: TextBlob(text).sentiment.polarity)
reviews["sentiment_category"] = reviews["sentiment"].apply(lambda score: "Positive" if score > 0 else "Negative" if score < 0 else "Neutral")

# Calculate overall sentiment:
average_sentiment = reviews["sentiment"].mean()
print("Overall sentiment:", average_sentiment)

# Count the number of positive, negative, and neutral reviews:
sentiment_counts = reviews["sentiment_category"].value_counts()
print("Sentiment distribution:\n", sentiment_counts)

# Access individual review sentiments:
for i in range(len(reviews)):
    print("Sentiment of the first review:", reviews["sentiment_category"].iloc[i])

# Further analysis and visualization:
# - Explore common themes in positive and negative reviews
# - Visualize sentiment distribution using charts
```
